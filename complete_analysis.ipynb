## a. Background and Context <a id="background-and-context"></a>

**Purpose**:

This research aims to explore how socioeconomic status and racial composition correlate with access to grocery stores in New York State. Specifically, the study will investigate the relationship between median income levels and the distance to the nearest grocery store across census tracts. It will also analyze how the racial composition of census tracts is associated with grocery store accessibility, identifying any disparities among racial groups. 

**Background Information**:

Access to nutritious and affordable food is a fundamental determinant of health and well-being. However, disparities in food access persist across socioeconomic and demographic lines, with significant implications for public health and social equity. Prior research suggests that such factors, including income and racial composition, play a role in determining food access (<a href="https://www.ers.usda.gov/amber-waves/2016/may/recent-evidence-on-the-effects-of-food-store-access-on-food-choice-and-diet-quality/#:~:text=People%20with%20poor%20access%20to,such%20as%20obesity%20or%20diabetes">USDA, 2016</a>). Low-income communities with a lack of access to food stores have to travel farther to shop for groceries or rely on stores with fewer choices or higher food prices. Poor access to foods such as fruits/vegetables, whole grains, lean meats, and low-fat dairy products could lead to poor diets and health conditions such as obesity or diabetes. Meanwhile, minority-majority neighborhoods frequently face systemic barriers to accessing well-stocked grocery stores, contributing to poorer dietary outcomes and exacerbating health disparities (<a href="https://www.nimhd.nih.gov/resources/understanding-health-disparities/food-accessibility-insecurity-and-health-outcomes.html">NIMHD, 2023</a>). The "food desert" concept, which highlights geographic areas with limited access to healthy food, has evolved to include a broader understanding of the structural and systemic barriers that perpetuate inequitable access (<a href="https://pubmed.ncbi.nlm.nih.gov/20462784/"> Walker et al., 2010</a>). These include historical patterns of racial segregation, economic disinvestment in certain neighborhoods, and discriminatory practices in urban planning. Despite federal efforts to improve food store access, such as the $500 million investment since 2011, gaps remain in understanding the spatial and socioeconomic dynamics of food accessibility (<a href="https://www.ers.usda.gov/amber-waves/2016/may/recent-evidence-on-the-effects-of-food-store-access-on-food-choice-and-diet-quality/#:~:text=People%20with%20poor%20access%20to,such%20as%20obesity%20or%20diabetes">USDA, 2016</a>).

Our research will focus on New York State, as it contains an interesting mix of rural and urban areas, with varying levels of income, as well as elevated levels of reported food insecurity compared to surrounding states (<a href="https://www.usnews.com/news/best-states/articles/states-with-the-highest-food-insufficiency-rates">US News, 2024</a>). In fact, in New York State, approximately one in four adults report experiencing food insecurity, which again, is closely associated with chronic diseases such as hypertension, diabetes, and heart disease (<a href="https://www.health.ny.gov/statistics/prevention/injury_prevention/information_for_action/docs/2023-12_ifa_report.pdf">NYSDOH, 2023</a>). These disparities are not evenly distributed; food insecurity is most pronounced in counties like the Bronx, Queens, and Herkimer, where access to grocery stores and nutritious food options is limited. We hope to explore the underlying racial and economic reasons behind grocery store access through analyzing the correlation between these variables and the distance to grocery stores in different New York State tracts.

**Potential Applications**:

By analyzing patterns of grocery store accessibility across census tracts in New York State, this study seeks to uncover the extent of inequities and their variations across different income brackets and racial groups. The findings will provide critical insights into food accessibility and potential inequalities across diverse communities, highlighting areas where barriers to healthy food access exist based on income level and racial composition.

These results hold significant implications for policymakers, urban planners, and public health officials striving to address food insecurity and promote health equity. Insights from this research could inform targeted interventions, such as incentives for grocery store development in underserved areas or support for transportation initiatives aimed at improving access to healthy food. Furthermore, non-profit organizations and community advocates can leverage these findings to develop data-driven strategies that promote food equity across New York State. The methodology and findings may also serve as a model for similar studies in other regions, contributing to a broader understanding of the socioeconomic and racial dimensions of food accessibility and informing efforts to reduce health disparities nationwide.

However, regardless of results, it is important to recognize that grocery store access is a complicated problem that cannot be fully captured by the scope of our project due to the multitude of factors at play. There are inherent limitations to our data and analysis, which will be expanded upon in the limitations section. Consequently, our project and research question is not aiming to identify a specific assertion regarding race/income/grocery store access, but just to see if there are any overarching or interesting trends that may have implications and inform future research.

## c. Key Terms and Data Sources <a id = "key-terms-and-data-sources"></a>

### Key Terms:

- Census Tract: A small, relatively permanent statistical subdivision of a county used by the US Census Bureau to analyze population and housing data.
- Median Income: The income level that divides a population into two equal groups, with half earning more and half earning less.
- Racial Demographic: The composition of a population based on racial categories. Here, it's the composition of a census tract.
- Majority-Majority: A demographic term indicating that the majority population in a region consists of a single racial or ethnic group.
- Majority-Minority: A demographic term referring to regions where minority racial or ethnic groups together make up the majority of the population.
- Grocery Store: A retail store that primarily sells food and household items, often used as an indicator of food access in a community.
- Food Insecurity: The state of being without reliable access to a sufficient quantity of affordable, nutritious food.
- Food Desert: Geographic areas with limited access to affordable and nutritious food, often in low-income or rural areas.
- Socioeconomic Status (SES): The social and economic position of a person or group, often measured by income, education, and occupation.
- Health Disparities: Differences in health outcomes and access to healthcare services across different population groups, often influenced by social, economic, and environmental factors.

## d. Summary of Findings <a id="summary"></a>

This analysis of New York State grocery store access patterns revealed two key relationships:

1. **Socioeconomic Impact:** Income shows a significant positive correlation with store distance (β = 0.5461), but this relationship varies dramatically by geography - from strong correlations in rural counties (Montgomery: R² = 0.73) to weak or no correlation in urban areas (Kings: R² = 0.00). This suggests that geographic location, rather than income alone, is the primary driver of access patterns.

2. **Racial Demographics:** The racial composition analysis explains 32.5% of access variance, with contrasting patterns across groups. White populations show increased distances (β = 0.3175), while Hispanic (β = -0.2983), Asian (β = -0.1718), and Black (β = -0.0634) populations show decreased distances, though these relationships are heavily influenced by urban-rural settlement patterns.

These findings indicate that while both income and racial composition significantly affect grocery store access, their impacts are secondary to and moderated by geographic location. Urban areas generally show more consistent access regardless of demographics, while rural areas face greater access challenges across all population groups. This suggests the need for location-specific interventions that consider the unique challenges of both rural and urban food access patterns, perhaps comparing urban versus rural differences and their demographics.

### Step 1: Retail Food Store Data Cleaning

To clean the retail food store data, we first loaded raw store data with 24,221 entries and 15 columns. Then, we selected only relevant columns: "Entity Name" and "Georeference" which we renamed to be lowercase. From there, we removed any rows with missing georeference data to handle missing data. We extracted latitude and longitude from georeference column, by splicing the georeference string, and stored as separate variables. Then, we dropped the original georeference column since it's no longer needed. The final format and outcome include the columns: store name, latitude, and longitude. This is saved as [cleaned_retail_stores.csv](https://github.com/peytonrsmith/INFO-2950-Final-Project/blob/main/cleaned_retail_stores.csv).

### Step 2: Census Data Cleaning

We retrieved data using Census API with specific variables: Median household income, population counts by race (White, Black, Asian, Hispanic), and census tract identifiers (county and census tract number). We then renamed columns to more intuitive names and converted numeric columns to float types. Then, we replaced census placeholder values (-666666666) with NaN and dropped rows with missing median income. In order to ensure each census tract is properly represented, we later realized that the census tract identifiers must be merged and thus created the variable "name" which is a string of census tract number followed by county name. We additionally created percentage variables for each racial/ethic group for ease of later analysis. This is saved as [cleaned_census_data.csv](https://github.com/peytonrsmith/INFO-2950-Final-Project/blob/main/cleaned_census_data.csv).

### Step 3: Spatial Data Processing

Then, we loaded the census tract shapefile (which contains the geometric boundaries of these tracts; each polygon in the .shp file represents a single census tract, for purpose of visualization), and converted coordinates to appropriate projection (EPSG:5070). From that, we calculated centroids for each census tract, and converted centroids to a latitude and longitude format to match the retail store data for merging. We created GeoDataFrames for both stores and tract centroid (using GeoPandas), and calculated distance from each tract centroid to nearest store, adding columns for nearest store name and its distance (in meters). This is saved as [census_tracts_with_nearest_store.csv](https://github.com/peytonrsmith/INFO-2950-Final-Project/blob/main/census_tracts_with_nearest_store.csv).

### Step 4: Final Data Merging

Finally, we merged the census data with spatial distance data by matching on tract identifiers (the combined tract number and county name string column). After merging, we dropped unnecessary columns (name, longitude, latitude, geometry). 

Final Dataset Features:
* Census tract demographics (median income, racial percentages)

# Part 3: Preregistration Statements <a id="preregistration-statements"></a>

### **Hypothesis One:** 
Areas with lower income have systematically reduced access to grocery stores, potentially creating food deserts that disproportionately affect economically disadvantaged communities.

Prior research has shown that food access inequality often follows socioeconomic lines, with lower-income neighborhoods having fewer grocery options and relying more heavily on convenience stores (Walker et al., 2010). An analysis of over 28,000 zip codes found that low-income neighborhoods have 25% fewer chain supermarkets compared to middle-income areas (Powell et al., 2007).

* $H_0$: There is no relationship between income level and grocery store access ($\beta_1 = 0$)
* $H_1$: Lower income areas have reduced odds of grocery store access ($\beta_1 < 0$)

Analysis will examine this relationship using:

* Log-transformed regression models to account for non-linear relationships in distance metrics
* County-level analysis to control for regional variations in development patterns
* Cross-validation to ensure model robustness and generalizability
* Alpha value of 0.05 is chosen to represent a 95% confidence level. If the p-value is less than 0.05, the results will be evaluated as statistically significant.

We believe this will be interesting because it evaluates whether the commonly assumed link between income and grocery store access holds across diverse regions and communities. Even if the results do not find a significant relationship (if the p-value is greater than the alpha value of 0.05), this lack of association would challenge prevailing assumptions, prompting a reevaluation of how food access inequalities are conceptualized and addressed.

If significant relationships (if the p-value is less than the alpha value of 0.05) are found, the findings will provide empirical evidence to support initiatives aimed at improving food access in low-income areas, such as incentivizing grocery store openings or enhancing public transportation to existing stores.
If no significant relationships are found, this could suggest that factors other than income, such as zoning laws, urban density, or cultural preferences, play a more prominent role in shaping food deserts. Such results could redirect research and policy focus toward these alternative explanations.


### **Hypothesis Two**: 
Racial and ethnic disparities in food access manifest through increased distances to grocery stores in predominantly minority communities, reflecting historical patterns of residential segregation and disinvestment.

Research has demonstrated that food access disparities often align with racial segregation patterns. A nationwide study of 28,050 zip codes found that predominantly Black neighborhoods have roughly half the number of chain supermarkets compared to predominantly White neighborhoods, while Hispanic neighborhoods have only a third as many (Powell et al., 2007). This disparity persists even after controlling for income levels, suggesting race-specific barriers beyond economic factors (Raja et al., 2008).

* $H_0$: There is no relationship between minority population percentage and distance to grocery stores ($\beta = 0$)
* $H_1$: Higher minority population percentages are associated with greater distances to grocery stores ($\beta > 0$)

Analysis will examine this relationship through:
* Multiple regression models controlling for income and population density
* Separate analyses for different racial/ethnic groups to identify group-specific patterns
* Alpha value of 0.05 is chosen to represent a 95% confidence level. If the p-value is less than 0.05, the results will be evaluated as statistically significant.

We believe this will be interesting because it moves beyond income-based explanations of food deserts, shedding light on structural factors that disproportionately affect minority populations, examining social justice issues.

If significant relationships (if the p-value is less than the alpha value of 0.05) are found, the findings will provide robust evidence of racial inequities in food access, emphasizing the need for targeted interventions in minority communities. If no significant relationships are found (if the p-value is greater than the alpha value of 0.05), this result would challenge the assumption that racial composition is independently associated with food access disparities and could prompt researchers to investigate alternative explanations.

```python

# Calculate centroids and ensure values in correct format 
tracts_gdf['centroid'] = tracts_gdf.geometry.centroid
tracts_gdf['GEOID'] = tracts_gdf['GEOID'].astype(str)
df['GEOID'] = df['GEOID'].astype(str)


# Merge data
tracts_gdf = tracts_gdf.merge(df, how="inner", on="GEOID")

# Categorize income and distance
tracts_gdf["income_category"] = pd.qcut(
    tracts_gdf["median_income"], q=3, labels=["Low", "Medium", "High"]
)
tracts_gdf["distance_category"] = pd.qcut(
    tracts_gdf["distance_to_nearest_store"], q=3, labels=["Low", "Medium", "High"]
)
tracts_gdf["interaction_category"] = tracts_gdf["distance_category"].astype(str) + "-" + tracts_gdf["income_category"].astype(str)


num_categories = len(tracts_gdf["interaction_category"].unique())
cols_bivar = ListedColormap(cm.get_cmap("tab20", num_categories).colors)

# Plot legend
fig, ax = plt.subplots(figsize=(4, 4))
categories_sorted = sorted(tracts_gdf["interaction_category"].unique())
for i, cat in enumerate(categories_sorted):
    x = i % 3
    y = i // 3
    ax.scatter(x, y, color=cols_bivar.colors[i], s=100)
ax.set_xticks(range(3))
ax.set_yticks(range(3))
ax.set_xlabel("Income Level")
ax.set_ylabel("Distance to Grocery Store")
ax.set_xticklabels(["Low", "Medium", "High"])
ax.set_yticklabels(["Low", "Medium", "High"])
ax.set_title("Income and Distance Legend")
ax.grid(False)
plt.show()

# Plot bivariate map
fig, ax = plt.subplots(figsize=(10, 8))
tracts_gdf.plot(
    column="interaction_category",
    cmap=cols_bivar,
    linewidth=0.5,
    edgecolor="gray",
    ax=ax,
)
ax.set_title("Distance to Grocery Stores and Income Levels in New York State")
plt.axis("off")
plt.show()
```


The colored regions show the interaction between income (Low, Medium, High) and store distance (Low, Medium, High) through a 3x3 color scheme. While our initial findings might suggest better access in minority communities (with negative correlations for Black, Hispanic, and Asian populations), this primarily reflects historical patterns of urban development and population distribution rather than truly better access. The shorter distances in areas with higher minority populations are largely a function of urban density rather than equitable food access. This finding has crucial implications for food justice initiatives. Despite shorter physical distances, urban minority communities may face other significant barriers to food access, including limited store options, higher prices, quality concerns, and time poverty. The strong geographic clustering effects we observed suggest that addressing food access disparities requires looking beyond simple distance metrics to consider factors like store quality, cultural appropriateness of food options, economic barriers, and transportation accessibility. Additionally, the positive correlation with white populations (β = 0.3175) largely reflects suburban and rural settlement patterns where greater distances might be offset by better transportation options and higher vehicle ownership rates. These nuanced findings suggest that policy interventions need to address not just physical distance but the broader ecosystem of food access, including economic barriers, transportation infrastructure, and quality of available options.

Our analysis revealed that food access in New York State is more complex than initially assumed. While both income and racial composition correlate with grocery store access, these relationships are primarily driven by geographic factors, particularly the urban-rural divide. This suggests that effective policy interventions must focus on geographic accessibility while considering the unique challenges faced by different demographic groups within their local contexts.

# Part 6: Limitations <a id="limitations"></a>

### 1. Urban-Rural Confounding Effects
The analysis doesn't fully separate urban and rural factors, including:
* Different shopping patterns in urban vs rural areas
* Varying store formats (supercenters vs small grocers)
* Distinct transportation challenges

**Impact:** May oversimplify the complex relationship between urbanization and food access, leading to potentially misguided one-size-fits-all policy recommendations.

### 2. Transportation Access Not Considered
* The study omits critical transportation variables:
* Public transit availability
* Vehicle ownership rates
* Road network quality
* Walking infrastructure
* Weather-related accessibility barriers*

**Impact:** Could significantly underestimate access challenges for transit-dependent populations and overstate food accessibility in car-centric areas.

### 3. Population Density Variations
Analysis doesn't fully account for:
* Neighborhood-level population concentrations
* Temporal population shifts
* Commercial vs residential zoning impacts
* Development patterns affecting store placement

**Impact:** May misrepresent per capita food access in areas with highly variable population density.

### 4. Unaccounted Socioeconomic Factors
Key variables not captured include:
* Income disparities
* Employment patterns
* Health outcomes
* Cultural food preferences

**Impact:** Limits understanding of how social and economic factors interact with physical food access.

### 5. Geographic Model Limitations
The models don't fully explain:
* Regional market variations
* Local policy differences
* Historical development patterns
* Natural geographic barriers

**Impact:** Reduces ability to make accurate predictions about food access patterns in different geographic contexts.

### 6. Limitations of Significance Metrics
* Binary Thresholding: Using 𝛼=0.05 oversimplifies results into "significant" or "not significant," ignoring nuances near the threshold.
* P-hacking: As mentioned in class, p-hacking is proven to be prevalent in academia. However, given how small our p-values are compared to 0.05, we do not believe that will be a concern.

**Impact:** May lead to overconfidence in statistical results while underestimating real-world complexities and practical significance.
